```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
df = pd.DataFrame({'WEIGHT': [15,18,12,10],
                   'PRICE':[1,3,2,5]},
                   index=['Orange','Apple','Banana','Grape'])
print(df)
```

# 1) Min-Max scaler
x_new = (x - x_min)/(x_max - x_min)
transform features by scaling each feture to a given range.
this scaler responds well if the standard deviation is small and when a distribution is not Gaussian. This Scaler is sensitive to outliers.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
df1 = pd.DataFrame(scaler.fit_transform(df),columns=['WEIGHT','PRICE'],index=['Orange','Apple','Banana','Grape'])
```

```python
ax = df.plot.scatter(x='WEIGHT',y='PRICE',color=['red','green','blue','yellow'],marker = '*',s=80,label='BEFORE SCALING')
df1.plot.scatter(x='WEIGHT',y='PRICE',color=['red','green','blue','yellow'],marker = 'o',s=60,label='AFTER SCALING',ax=ax)
```

# 2) standard scaler

the standard scaler assumes that data is normally distributed within each feature and scales them such that the distribution centered around 0.
x_new = (x-mu)/stigma. if data is not normally distributed, this is not the best scaler to use.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
df2=pd.DataFrame(scaler.fit_transform(df),columns=['WEIGHT','PRICE'],index=['Orange','Apple','Banana','Grape'])
ax = df.plot.scatter(x='WEIGHT',y='PRICE',color=['red','green','blue','yellow'],marker='*',s=80,label='BEFORE scaLing')
df2.plot.scatter(x='WEIGHT',y='PRICE',color=['red','green','blue','yellow'],marker = 'o',s=60,label='AFTER SCALING',ax=ax)
plt.axhline(0,color='red',alpha=0.2)
```
