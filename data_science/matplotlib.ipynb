```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0,6])
y = np.array([0,8])
```

```python
plt.plot(x,y)
plt.show()
```

# Line Chart

```python
points = np.array([4,7,2,10,13])
plt.plot(points)
plt.show()
```

```python
a2 = np.array([1,2,3,4,5,6])
plt.plot(a2,linestyle='dotted')
plt.show()
```

# Bar Chart

```python
x = np.array(["X","Y","Z","W"])
y = np.array([25,30,27,22])
plt.bar(x,y)
plt.show()
```

# Scatter Plot

```python
x = np.array([2,7,15,8,10,12])
y = np.array([25,30,27,22,8,10])
plt.scatter(x,y)
plt.show()
```

# Pie Chart

```python
x = np.array([30,20,25,20])
plt.pie(x)
plt.show()
```

# Histogram

```python
x = np.random.normal(10,5,250)
y = np.array([1,2,3,4])
print(x)
plt.hist(x)
plt.show()
```

## Distplot

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
age1 = titanic['age'].dropna()
sns.displot(age1,bins=30,kde=False)
plt.show
```

## Regplot

```python
data = sns.load_dataset('mpg')
```

```python
data.head()
```

```python
sns.regplot(x='mpg',y='acceleration',data=data)
plt.show()
```
